```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
trialdata_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_df, trialdata_df, how='outer', on="Mouse ID")

# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Using GroupBy in order to separate the data into fields according to "Drug" and "Timepoint"
grouped_tumor_df = combined_df.groupby(['Drug','Timepoint'])

# Calculate the Mean Tumor Volume
tumor_mean_df = pd.DataFrame(grouped_tumor_df['Tumor Volume (mm3)'].mean())

# Preview DataFrame
tumormean_df = tumor_mean_df.reset_index()
tumormean_df.head()
```

```python
# Minor Data Munging to Re-Format the Mean Tumor Value Data Frames
#Using pivot() to convert a column to row name/index in dataframe.
format_tumormean_df = tumormean_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)') 
format_tumormean_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Calculate standard error on means
tumor_sem_df = pd.DataFrame(grouped_tumor_df['Tumor Volume (mm3)'].sem())

# Preview DataFrame
tumorsem_df = tumor_sem_df.reset_index()
tumorsem_df.head()
```

```python
#Using pivot() to convert a column to row name/index in dataframe.
format_tumorsem_df = tumorsem_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)') 
format_tumorsem_df
```

```python
# Generate the Plot (with Error Bars)
#Setting up the plot - plt.errorbar(x_axis,mean,yerr=sem, color, fmt="o", linestyle)
x_axis = np.arange(0, 50, 5)
plt.ylim(35,75)

plt.errorbar(x_axis,format_tumormean_df['Capomulin'],yerr=format_tumorsem_df['Capomulin'],
             color="red", fmt="o", linestyle='--', label='Capomulin')
plt.errorbar(x_axis,format_tumormean_df['Infubinol'],yerr=format_tumorsem_df['Infubinol'],
             color="blue", fmt="o", linestyle='--', label='Infubinol')
plt.errorbar(x_axis,format_tumormean_df['Ketapril'],yerr=format_tumorsem_df['Ketapril'],
             color="green", fmt="o", linestyle='--', label='Ketapril')
plt.errorbar(x_axis,format_tumormean_df['Placebo'],yerr=format_tumorsem_df['Placebo'], 
             color="black", fmt="o", linestyle='--', label='Placebo')

# Lables
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc='upper left')
plt.grid()

# Save the Figure
plt.savefig ("Tumor Response.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Calculate the Mean Metastatic Sites
meta_mean_df = (grouped_tumor_df['Metastatic Sites'].mean())
# Convert to DataFrame
meta_mean_df = pd.DataFrame(grouped_tumor_df['Metastatic Sites'].mean())
# Preview DataFrame
metamean_df = meta_mean_df.reset_index()
metamean_df.head()
```

```python
#Using pivot() to convert a column to row name/index in dataframe.
format_metamean_df = metamean_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites') 
# Preview that Reformatting worked
format_metamean_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_sem_df = (grouped_tumor_df['Metastatic Sites'].sem())
# Convert to DataFrame
meta_sem_df = pd.DataFrame(grouped_tumor_df['Metastatic Sites'].sem())
# Preview DataFrame
metasem_df = meta_sem_df.reset_index()
metasem_df.head()
```

```python
# Minor Data Munging to Re-Format the Standard Error Met. Sites Data Frame
#Using pivot() to convert a column to row name/index in dataframe.
format_metasem_df = metasem_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites') 
# Preview that Reformatting worked
format_metasem_df
```

```python
# Generate the Plot (with Error Bars)
#Setting up the plot - plt.errorbar(x_axis,mean,yerr=sem, color, fmt="o", linestyle)
x_axis = np.arange(0, 50, 5)
plt.ylim(0,3.5)

plt.errorbar(x_axis,format_metamean_df['Capomulin'],yerr=format_metasem_df['Capomulin'],
             color="red", fmt="o", linestyle='--', label='Capomulin')
plt.errorbar(x_axis,format_metamean_df['Infubinol'],yerr=format_metasem_df['Infubinol'],
             color="blue", fmt="o", linestyle='--', label='Infubinol')
plt.errorbar(x_axis,format_metamean_df['Ketapril'],yerr=format_metasem_df['Ketapril'],
             color="green", fmt="o", linestyle='--', label='Ketapril')
plt.errorbar(x_axis,format_metamean_df['Placebo'],yerr=format_metasem_df['Placebo'], 
             color="black", fmt="o", linestyle='--', label='Placebo')

# Lables
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met.Sites")
plt.legend(loc='upper left')
plt.grid()

# Save the Figure
plt.savefig ("Metastatic Spread.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count_df = pd.DataFrame(grouped_tumor_df['Mouse ID'].count())

# Convert to DataFrame
mouse_df=mouse_count_df.reset_index()

# Preview DataFrame
mouse_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#Using pivot() to convert a column to row name/index in dataframe.
format_mouse_df = mouse_df.pivot(index='Timepoint', columns='Drug', values='Mouse ID') 
# Preview the Data Frame
format_mouse_df
```

```python
# Generate the Plot (Accounting for percentages)

x_axis = np.arange(0, 50, 5)
plt.ylim(30,110)

plt.errorbar(x_axis,format_mouse_df['Capomulin']/25*100, color="red", fmt="o", linestyle='--', label='Capomulin')
plt.errorbar(x_axis,format_mouse_df['Infubinol']/25*100, color="blue", fmt="o", linestyle='--', label='Infubinol')
plt.errorbar(x_axis,format_mouse_df['Ketapril']/25*100, color="green", fmt="o", linestyle='--', label='Ketapril')
plt.errorbar(x_axis,format_mouse_df['Placebo']/25*100, color="black", fmt="o", linestyle='--', label='Placebo')

# Lables
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rates (%)")
plt.legend(loc='lower left')
plt.grid()

# Save the Figure
plt.savefig ("Survival During Treatment.png")
# Show the Figure
plt.show()
```
